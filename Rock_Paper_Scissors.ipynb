```python
import random
```

```python
def player(prev_play, opponent_history=[]):
    if prev_play == "":
        return random.choice(["R", "P", "S"])
    else:
        return random.choice(["R", "P", "S"])
```

```python
def quincy(prev_play, opponent_history=[]):
    return "R"

def mrugesh(prev_play, opponent_history=[]):
    return "P"

def abbey(prev_play, opponent_history=[]):
    return "S"

def kris(prev_play, opponent_history=[]):
    return random.choice(["R", "P", "S"])
```

```python
def play(player1, player2, num_games, verbose=False):
    p1_wins = 0
    p2_wins = 0
    ties = 0
    for i in range(num_games):
        p1 = player1("", [])
        p2 = player2("", [])
        if p1 == p2:
            ties += 1
        elif (p1 == "R" and p2 == "S") or (p1 == "S" and p2 == "P") or (p1 == "P" and p2 == "R"):
            p1_wins += 1
        else:
            p2_wins += 1
        if verbose:
            print(f"Game {i+1}: Player 1: {p1}, Player 2: {p2}")
    if verbose:
        print("Results:")
        print(f"Player 1 wins: {p1_wins}")
        print(f"Player 2 wins: {p2_wins}")
        print(f"Ties: {ties}")
    return p1_wins, p2_wins, ties

# Test your player against one of the opponents
play(player, quincy, 1000, verbose=True)
```
